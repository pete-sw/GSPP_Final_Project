```python
#Mapping imports
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
%pylab inline
from pylab import *
import numpy as np

# Set default figure size for this notebook
pylab.rcParams['figure.figsize'] = (12, 8)
```

```python
#Generate poisson lats and poisson lons, zip them together

#In case I need to map to different scale
#from scipy.interpolate import interp1d
#m = interp1d([1,512],[5,10])

#They initially need to fit in the US Bounding Box: -125.0011, 24.9493, -66.9326, 49.5904, Centroid -95.9669, 37.1669
lat_PP = np.random.poisson(95.9669, 300)
lon_PP = np.random.poisson(37.1669, 300)
lat_lon_PP = zip(-lat_PP, lon_PP)
lat_lon_PP = [list(elem) for elem in lat_lon_PP]
lat_lon_PP
```

```python
#Problem with this is that the points aren't constrained by the US outline
import shapefile
sf = shapefile.Reader("../Data/states_21basic/lower_48")
us_shapes = sf.shapes() #49 shapes now, all 48 states and DC, including a few islands
#us_shapes[1].points
#len(us_shapes)
#The points of each shape can be treated as a polygon, which is just a list of x/y pairs
#then using the next function to see if points are inside
```

```python
#Now let's define a function to see if our Poisson points are inside of these polygons
def point_inside_polygon(x,y,poly):

    n = len(poly)
    inside =False

    p1x,p1y = poly[0]
    for i in range(n+1):
        p2x,p2y = poly[i % n]
        if y > min(p1y,p2y):
            if y <= max(p1y,p2y):
                if x <= max(p1x,p2x):
                    if p1y != p2y:
                        xinters = (y-p1y)*(p2x-p1x)/(p2y-p1y)+p1x
                    if p1x == p2x or x <= xinters:
                        inside = not inside
        p1x,p1y = p2x,p2y

    return inside
```

```python
#And let's apply this function to our Poisson points and the 49 polygons we now have
inside_lat_lon_PP = []
for x in lat_lon_PP:
    for y in range(49):
        if point_inside_polygon(x[0], x[1], us_shapes[y].points):
            inside_lat_lon_PP.append(x)
inside_lat_lon_PP
```

```python
lon_ppoints = []
lat_ppoints = []

#Iterate through our list of lists and split out lat and long into their own lists for the mapping
for x in inside_lat_lon_PP:
    lon_ppoints.append(int(x[0]))
    lat_ppoints.append(int(x[1]))

#Define map layout
map = Basemap(projection='merc', lat_0 = 40, lon_0 = -80,
    resolution = 'l', area_thresh = 9000,
    llcrnrlon=-125, llcrnrlat=24,
    urcrnrlon=-67, urcrnrlat=49)
map.drawcoastlines()
map.drawcountries()
map.fillcontinents(color = 'gray')
map.drawmapboundary()
   
#Put points in
for lon, lat in zip(lon_ppoints, lat_ppoints):
    x,y = map(lon_ppoints, lat_ppoints)
    map.plot(x, y, 'ko')
plt.show()
```
