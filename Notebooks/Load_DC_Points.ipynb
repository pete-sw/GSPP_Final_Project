### Initial Map

```python
#Getting started on incorporating the Pandas data to this map
#Colors
categories = np.unique(df['Company'])
#categories
colors = ['#00933B', '#F2B50F', '#F90101', '#0266C8', '#0266C8', '#0266C8']
colordict = dict(zip(categories, colors))  
colordict
#color_dict = {'Google': '#00933B', 'Microsoft': '#F2B50F', 'Equinix': '#F90101', 'CoreSite': '#0266C8', 'Sabey': '#0266C8'}
df['Color2'] = df['Company'].apply(lambda x: colordict[x])
df.head()
```

```python
#Get the values out of the dataframe that we want
lon = df['lon'].tolist()
lat = df['lat'].tolist()
colors = df.Color2.tolist()

x,y = map(lon, lat)
# map = Basemap(projection='merc',lat_0=40,lon_0=-80,resolution='l', llcrnrlon=-125, llcrnrlat=24,
#     urcrnrlon=-67, urcrnrlat=49)

map = Basemap(projection='merc', lat_0 = 40, lon_0 = -80,
    resolution = 'l', area_thresh = 0.1,
    llcrnrlon=-125, llcrnrlat=24,
    urcrnrlon=-67, urcrnrlat=49)

map.drawcoastlines()
map.drawcountries()

#map.fillcontinents(color = 'gray')
map.bluemarble()
map.drawmapboundary()

for lon, lat, dot_color in zip(lon, lat, colors):
    x,y = map(lon, lat)
    map.plot(x, y, 'o', color=dot_color)
plt.show()
```
