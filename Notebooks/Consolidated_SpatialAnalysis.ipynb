```python
#Mapping imports
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
%pylab inline
from pylab import *
import numpy as np

# Set default figure size for this notebook
pylab.rcParams['figure.figsize'] = (12, 8)
```

### Point Generator Functions (Poisson vs Random)

###### Poisson

```python
#Poisson- may not be right because it should be for generating # events, not location of an event?
#They initially need to fit in the US Bounding Box: -125.0011, 24.9493, -66.9326, 49.5904, Centroid -95.9669, 37.1669
#Generate poisson lats and poisson lons, zip them together
def poisson_generator(points_to_generate):
    #Need to deal with fact that np.random.poisson only returns integers
    lat_PP = np.random.poisson(95.9669, points_to_generate).astype(float)
    for index, coords in enumerate(lat_PP):
        jitter =  uniform(0, 1)
        lat_PP[index] = coords+jitter
    lon_PP = np.random.poisson(37.1669, points_to_generate).astype(float)
    for index, coords in enumerate(lon_PP):
        jitter =  uniform(0, 1)
        lon_PP[index] = coords+jitter
    lat_lon_PP = zip(-lat_PP, lon_PP)
    #They come in as tuples from the zip so we need to make them a list of lists
    lat_lon_PP = [list(elem) for elem in lat_lon_PP]
    return lat_lon_PP
```

###### Random

```python
def random_generator(points_to_generate):
    lats = []
    lons = []
    for x in range(points_to_generate):
        lats.append(uniform(24.9493, 49.5904))
    for x in range(points_to_generate):
        lons.append(uniform(-125.0011, -66.9326))
    lat_lon_PP = zip(lons, lats)
    #They come in as tuples from the zip so we need to make them a list of lists
    lat_lon_PP = [list(elem) for elem in lat_lon_PP]
    return lat_lon_PP
```

### Let's map these Random and Poisson points

```python
#Let's define a function to see if our random points are inside of these polygons

def point_inside_polygon(x,y,poly):
    n = len(poly)
    inside =False
    p1x,p1y = poly[0]
    for i in range(n+1):
        p2x,p2y = poly[i % n]
        if y > min(p1y,p2y):
            if y <= max(p1y,p2y):
                if x <= max(p1x,p2x):
                    if p1y != p2y:
                        xinters = (y-p1y)*(p2x-p1x)/(p2y-p1y)+p1x
                    if p1x == p2x or x <= xinters:
                        inside = not inside
        p1x,p1y = p2x,p2y
    return inside
```

```python
#This function returns a list of lat/long lists
#It has n points (as defined by count)
#and can be poisson or random

def generate_points(dist_type, count):
    inside_lat_lon_PP = []
    points_generated = 0
    total_to_generate = count
    #Can set however many points I want to generate up here
    #it will create that many randomly generated poisson points inside the US shapefile
    while points_generated < total_to_generate:

        #Adding one because the while loop keeps being a jerk, not sure why but it breaks at n-1
        points_to_generate = (total_to_generate - points_generated)+1
        #lat_lon_PP = poisson_generator(points_to_generate)
        if dist_type == 'random':
            lat_lon_PP = random_generator(points_to_generate)
        elif dist_type == 'poisson':
            lat_lon_PP = poisson_generator(points_to_generate)
        else:
            print "Please enter random or poisson"
            break
        #Let's do a quick check to just make sure the points are in the right range, was having a few weird outliers
        for index, value in enumerate(lat_lon_PP):
            if lat_lon_PP[index][0] < 0:
                lat_lon_PP.remove(value)

        #And let's apply the inside function to our Poisson points and the 49 polygons we loaded
        temp_inside_lat_lon_PP = []
        for x in lat_lon_PP:
            for y in range(49):
                #https://code.google.com/p/pyshp/wiki/PyShpDocs
                if point_inside_polygon(x[0], x[1], us_shapes[y].points):
                    temp_inside_lat_lon_PP.append(x)


        #Keep track of how many we have created so far
        points_generated += len(temp_inside_lat_lon_PP)
        #Append the points we've generated to a final list
        inside_lat_lon_PP.extend(temp_inside_lat_lon_PP)
    return inside_lat_lon_PP
```

```python
#And now let's define a function to map the results

def map_points(point_list):
    lon_ppoints = []
    lat_ppoints = []

    #Iterate through our list of lists and split out lat and long into their own lists for the mapping
    for x in point_list:
        lon_ppoints.append(int(x[0]))
        lat_ppoints.append(int(x[1]))
    #Define map layout
    map = Basemap(projection='merc', lat_0 = 40, lon_0 = -80,
        resolution = 'l', area_thresh = 9000,
        llcrnrlon=-125, llcrnrlat=24,
        urcrnrlon=-67, urcrnrlat=49)
    map.drawcoastlines()
    map.drawcountries()
    map.fillcontinents(color = 'gray')
    #map.drawmapboundary()

    #Put points in
    for lon, lat in zip(lon_ppoints, lat_ppoints):
        x,y = map(lon_ppoints, lat_ppoints)
        map.plot(x, y, 'ko')
    plt.show()
```

# Distance to All Other Points- Poisson, Random and Actual Data Centers

```python
#This function actually gets the right distance for lon/lat, because Euclidean distance would be inaccurate
#from http://www.johndcook.com/blog/python_longitude_latitude/
#multiply by 3960 for value in miles
#multiply by 6373 for value in km

import math
 
def distance_on_unit_sphere(lat1, long1, lat2, long2):
 
    # Convert latitude and longitude to 
    # spherical coordinates in radians.
    degrees_to_radians = math.pi/180.0
         
    # phi = 90 - latitude
    phi1 = (90.0 - lat1)*degrees_to_radians
    phi2 = (90.0 - lat2)*degrees_to_radians
         
    # theta = longitude
    theta1 = long1*degrees_to_radians
    theta2 = long2*degrees_to_radians
         
    # Compute spherical distance from spherical coordinates.
         
    # For two locations in spherical coordinates 
    # (1, theta, phi) and (1, theta, phi)
    # cosine( arc length ) = 
    #    sin phi sin phi' cos(theta-theta') + cos phi cos phi'
    # distance = rho * arc length
     
    cos = (math.sin(phi1)*math.sin(phi2)*math.cos(theta1 - theta2) + 
           math.cos(phi1)*math.cos(phi2))
    arc = math.acos( cos )
 
    # Remember to multiply arc by the radius of the earth 
    # in your favorite set of units to get length.
    return arc
```

```python
#Create a distance matrix from whatever lat_long pairs you give it

def create_dist_matrix(lat_lon_pairs):
    dist_matrix = np.zeros(shape=(len(lat_lon_pairs), len(lat_lon_pairs)))
    for index, value in enumerate(lat_lon_pairs):
        for secondary_index, secondary_value in enumerate(lat_lon_pairs):
    #         print "Index, ", index, value
    #         print "Secondary", secondary_index, secondary_value
            #print "Index= " + str(index) + ', Val= ' + str(value)
            #for secondary_index in range(len(dist_matrix[0])):
            #print dist_matrix[index][secondary_index]
            #print "Dist_Matrix= " + str(dist_matrix[index][secondary_index])

            #Getting a stupid value error where it can't calculate distance between same points that should be zero
            while True:
                try:
                    #print "No problem at, ", index, secondary_index
                    dist_matrix[index][secondary_index] = 3960*distance_on_unit_sphere(value[0], value[1], secondary_value[0], secondary_value[1])
                    break
                except ValueError:
                    #print "Problem! ", index, secondary_index
                    dist_matrix[index][secondary_index] = 0
                    break

            if dist_matrix[index][secondary_index] < .00001:
                dist_matrix[index][secondary_index] = 0
    return dist_matrix
```

# Nearest neighbor analysis for Random, Poisson and Actual Data Center Locations

```python
#Nearest neighbor function
def nearest_neighbor(input_array):
    #Create a nearest neighbor list
    nearest_neighbor_dist = []
    #For each data center (using column as the index)
    for index, row in enumerate(input_array):
        #Find the minimum value that is greater than (about) zero
        nearest_neighbor_dist.append(min(x for x in row if x > .0001))
        #print min(row)
    return nearest_neighbor_dist
```

### What about not Nearest Neighbor, but neighbor N neighbors away?

```python
#Select the nth smallest value in a list
#from http://code.activestate.com/recipes/269554-select-the-nth-smallest-element/
import random
def select_smallest(data, n):
    "Find the nth rank ordered element (the least value has rank 0)."
    data = list(data)
    if not 0 <= n < len(data):
        raise ValueError('not enough elements for the given rank')
    while True:
        pivot = random.choice(data)
        pcount = 0
        under, over = [], []
        uappend, oappend = under.append, over.append
        for elem in data:
            if elem < pivot:
                uappend(elem)
            elif elem > pivot:
                oappend(elem)
            else:
                pcount += 1
        if n < len(under):
            data = under
        elif n < len(under) + pcount:
            return pivot
        else:
            data = over
            n -= len(under) + pcount
```

```python
#Nearest N neighbor function
def n_nearest_neighbor(input_array, nth_neighbor):
    #Create a nearest neighbor list
    n_nearest_neighbor_dist = []
    #For each data center (using column as the index)
    for index, row in enumerate(input_array):
        #Find the N minimum value
        n_nearest_neighbor_dist.append(select_smallest(row, nth_neighbor))
        #print row
    return n_nearest_neighbor_dist
```

### Algorithmic Clustering

```python
#see #http://geoffboeing.com/2014/08/clustering-to-reduce-spatial-data-set-size/

from sklearn.cluster import DBSCAN
from geopy.distance import great_circle
#Get our data out again in case things have changed

coordinates = dc_df.as_matrix(columns=['lon', 'lat'])
```

```python
#Set up some functions that DBSCAN needs
def getCentroid(points):
    n = points.shape[0]
    sum_lon = np.sum(points[:, 1])
    sum_lat = np.sum(points[:, 0])
    return (sum_lon/n, sum_lat/n)

def getNearestPoint(set_of_points, point_of_reference):
    closest_point = None
    closest_dist = None
    for point in set_of_points:
        point = (point[1], point[0])
        dist = great_circle(point_of_reference, point).meters
        if (closest_dist is None) or (dist < closest_dist):
            closest_point = point
            closest_dist = dist
    return closest_point
```

```python
#Enter the coordinates that you want to submit and the minimum # of samples to define a cluster
def plot_DBSCAN_clusters(coordinates, min_samples):
    #Can play with epsilon value (distance betwee datacenters to consider) as eps = .01 or .05
    #Can play with minimum samples
    db = DBSCAN(min_samples=min_samples).fit(coordinates)
    labels = db.labels_
    num_clusters = len(set(labels)) - (1 if -1 in labels else 0)
    clusters = pd.Series([coordinates[labels == i] for i in xrange(num_clusters)])
    print('Number of clusters: %d' % num_clusters)
    
    lon = []
    lat = []
    for i, cluster in clusters.iteritems():
        if len(cluster) < 3:
            representative_point = (cluster[0][1], cluster[0][0])
        else:
            representative_point = getNearestPoint(cluster, getCentroid(cluster))
        lon.append(representative_point[0])
        lat.append(representative_point[1])
    rs = pd.DataFrame({'lon':lat, 'lat':lon})
    
    plt.figure(figsize=(14, 9), dpi=100)

    map = Basemap(projection='merc', lat_0 = 40, lon_0 = -80,
        resolution = 'l', area_thresh = 9000,
        llcrnrlon=-130, llcrnrlat=24,
        urcrnrlon=-67, urcrnrlat=49)
    map.drawcoastlines()
    map.drawcountries()


    #map.fillcontinents(color = 'gray')

    for lon_DC_RS, lat_DC_RS in zip(rs['lon'], rs['lat']):
        x,y = map(lon_DC_RS, lat_DC_RS)
        #map.plot(x, y, 'r*')
        map.scatter(x, y, c='b', alpha=.4, s=400, label='Cluster')

    for lon_DC_DF, lat_DC_DF in zip(dc_df['lon'], dc_df['lat']):
        x,y = map(lon_DC_DF, lat_DC_DF)
        #map.plot(x, y, 'r*')
        map.scatter(x, y, c='r', alpha=.5, s=20, label='Data Center')
    plt.axis('off')
    #plt.legend(loc='upper right', numpoints = 1)
    plt.title('Minimum Cluster Size: '+ str(min_samples), size='18')
    #plt.legend((dc_df_scatter, rs_scatter), ('All Data Centers', 'Clusters'), loc='best')
```
