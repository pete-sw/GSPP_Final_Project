```python
import shapefile
sf = shapefile.Reader("Data/gregs_cable_map_v1.53/Cables.shp")
shapes = sf.shapes()
# Read the bounding box from the 4th shape
shapes[3].bbox
```

```python
#Mapping imports
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize=(15.7,12.3))
ax = plt.subplot(111)

map = Basemap(projection='cyl',llcrnrlat=-90,urcrnrlat=90,\
            llcrnrlon=-90,urcrnrlon=90,resolution='c')

map.drawmapboundary(fill_color='aqua')

map.readshapefile('Data/gregs_cable_map_v1.53/Cables', 'Cables', drawbounds=True)
```
