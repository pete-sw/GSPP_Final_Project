```python
import geocoder
```

```python
counter = 0
with open('Data/exchange_addresses.txt') as f:
    content = f.readlines()
    #lines = (line.rstrip('\n') for line in open('Data/exchange_addresses.txt'))
for item in content:
    item = item.strip('\n')
    content[counter] = item
    counter += 1
#print content
```

```python
lat_long = []
for address in content:
    temp = geocoder.google("<"+str(address)+">")
    lat_long.append([temp.lat, temp.lng])
len(lat_long)
```

```python
%pylab inline
from pylab import *

# Set default figure size for this notebook
pylab.rcParams['figure.figsize'] = (8.0, 6.4)
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
```

```python
map = Basemap(projection='merc', lat_0 = 40, lon_0 = -80,
    resolution = 'l', area_thresh = 0.1,
    llcrnrlon=-125, llcrnrlat=24,
    urcrnrlon=-67, urcrnrlat=49)

map.drawcoastlines()
map.drawcountries()

#map.fillcontinents(color = 'gray')
map.bluemarble()
map.drawmapboundary()

lat = []
lon = []

for x in lat_long:
    lat.append(x[0])
    lon.append(x[1])

for lon, lat in zip(lon, lat):
    x,y = map(lon, lat)
    map.plot(x, y, 'ko')
plt.show()
```
