# Loading and mapping all the datasets together

### Data centers

```python
%pylab inline
from pylab import *

# Set default figure size for this notebook
pylab.rcParams['figure.figsize'] = (12, 8)
```

```python
#Load data from CSV
import pandas as pd
#df = df.dropna()
df = pd.DataFrame.from_csv('data/Initial_DC_Locations.csv')
for x in range(len(df['SQFT'])):
    df['SQFT'][x] = df['SQFT'][x].replace(',', '')
df[['lon', 'lat', 'SQFT']] = df[['lon', 'lat', 'SQFT']].astype(float)
df = df.reset_index()
df.head()
```

```python
#Mapping imports
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Getting started on incorporating the Pandas data to this map
#Colors
categories = np.unique(df['Company'])
#categories
colors = ['#00933B', '#F2B50F', '#F90101', '#0266C8', '#0266C8', '#0266C8']
colordict = dict(zip(categories, colors))  
colordict
#color_dict = {'Google': '#00933B', 'Microsoft': '#F2B50F', 'Equinix': '#F90101', 'CoreSite': '#0266C8', 'Sabey': '#0266C8'}
df['Color2'] = df['Company'].apply(lambda x: colordict[x])
df.head()
```

```python
#Get the values out of the dataframe that we want
lon_DC = df['lon'].tolist()
lat_DC = df['lat'].tolist()
colors = df.Color2.tolist()


# map = Basemap(projection='merc',lat_0=40,lon_0=-80,resolution='l', llcrnrlon=-125, llcrnrlat=24,
#     urcrnrlon=-67, urcrnrlat=49)

map = Basemap(projection='merc', lat_0 = 40, lon_0 = -80,
    resolution = 'l', area_thresh = 0.1,
    llcrnrlon=-125, llcrnrlat=24,
    urcrnrlon=-67, urcrnrlat=49)

x,y = map(lon_DC, lat_DC)

map.drawcoastlines()
map.drawcountries()

#map.fillcontinents(color = 'gray')
map.bluemarble()
map.drawmapboundary()

for lon_DC, lat_DC, dot_color in zip(lon_DC, lat_DC, colors):
    x,y = map(lon_DC, lat_DC)
    map.plot(x, y, 'o', color=dot_color)
plt.show()
```

### Internet Exchanges

```python
import geocoder
```

```python
counter = 0
with open('Data/exchange_addresses.txt') as f:
    content = f.readlines()
    #lines = (line.rstrip('\n') for line in open('Data/exchange_addresses.txt'))
for item in content:
    item = item.strip('\n')
    content[counter] = item
    counter += 1
#print content
lat_long = []
for address in content:
    temp = geocoder.google("<"+str(address)+">")
    lat_long.append([temp.lat, temp.lng])
len(lat_long)
```

```python
lat_IX = []
lon_IX = []

for x in lat_long:
    lat_IX.append(x[0])
    lon_IX.append(x[1])

for lon, lat in zip(lon_IX, lat_IX):
    x,y = map(lon_IX, lat_IX)
    map.plot(x, y, 'ko')
plt.show()
```
